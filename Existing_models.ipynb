```python
import re
def clean_tweet_text(tweet):
    text = re.sub(r'@\w+\s?', '', tweet)
    text = re.sub(r'http.?://[^\s]+[\s]?', '', text)
#     text = re.sub(r'#\w+\s?', '', tweet)
    text = re.sub('[^a-zA-Z\s]\s?', '', text)
    text = text.lower()
    return text
```

```python
test = "RT @handle: Director of Global Brand Marketing, Hotels and Casino's $125k + 30% bonus - Orlando Fl http://bit.ly/4kUmBB #jobs #twitjobs"
clean_test = clean_tweet_text(test)
print (clean_test)
```
